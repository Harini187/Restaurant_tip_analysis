```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# --- Task 1: Your First Matplotlib Chart ---
plt.figure(figsize=(8, 5))
daily_tips = tips.groupby("day", observed=True)["tip"].sum()

plt.bar(daily_tips.index, daily_tips.values, color='skyblue')
plt.title("Total Tip Amount Collected by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Sum of Tips ($)")
plt.show()

# --- Task 2: Seaborn Histogram ---
plt.figure(figsize=(8, 5))
sns.histplot(tips['tip'], bins=20, kde=True, color='green')
plt.title("Distribution of Tip Amounts")
plt.show()
```

Most tips are between $2 and $4, and they get less common as the price goes up. The chart is right-skewed because a few big tips pull the average higher. The KDE line smooths out the bars, making it easier to see where most people usually tip.

```python
# --- Task 3: Seaborn Scatter Plot ---
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip Amount")
plt.show()
```

The chart shows that as the total bill goes up, the tip generally goes up too. You can also see that people tend to tip more consistently on certain days, like weekends.
